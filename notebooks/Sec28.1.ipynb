```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv("forest.csv",sep="\t",header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype('int')
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Remove null values
df=df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
df.iloc[[122]]
```

```python
df.columns
```

```python
# fix spaces in column name
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype('int')
```

```python
df.info()
```

```python
# changing the other columns to float datatype
objects=[features for features in df.columns if df[features].dtype=='O']#O stands for object datatype
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype('float')
```

```python
df.info()
```

```python
objects
```

```python
df.describe()
```

```python
# let's save the cleaned data
df.to_csv("forest_cleaned.csv",index=False)
```

```python
# EDA
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
# Encoding of the categories in classes
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
# Categories in classes after encoding
df_copy['Classes'].value_counts()
```

```python
# plot density plot for all features
plt.style.use('seaborn-v0_8-darkgrid')
df_copy.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
# percentage for pie chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
# plotting piechart
classlabels=['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Distribution of Fire and Not Fire")
plt.show()
```

```python
# correlation
df_copy.corr()
```

```python
sns.heatmap(df_copy.corr(),annot=True,cmap='coolwarm')
plt.show()
```
